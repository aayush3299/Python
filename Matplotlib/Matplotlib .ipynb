## Matplotlib 
<b>is a plotting library</b> <br>
In simple words it is a library which helps you create high level graphs (much more sophisticated than excel ones).<br>

```python
#importing matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
#tells jupyter to print out of plot without specifying plt.show()
```

```python
x=np.arange(0,10)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
```

```python
yy=np.arange(0,20,0.01)
z=np.cos(yy)
plt.plot(yy,z,'-r')
#'-r' stands for color argument
plt.xlabel('number')
plt.ylabel('cos')
plt.title('Cos vs Num Graph')
```

<hr><code><b>Subplot</b>(number of rows,number of columns,position of graph)</code> is used to divide plane into multiple parts .<br>
first arg stands for number of rows<br>
second for number of columns<br>
third arg stands for position of that particular graph in the plane.<br>


```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

<hr><code><b>add_axes</b>(left,bottom,width,height)</code><br>
It is another way to add subplots to an plot.<br>
It has more options like width and such

```python
fig=plt.figure()
#creates a named figure

axes=fig.add_axes([0.1,0.1,0.8,0.8])
#the decimal 0.1 stands for 10%
#hence the first 2 args mean 10%from left and 10% from border

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0,0,1,1])
axes1.set_title('Larger axis')

axes2=fig.add_axes([0.3,0.5,0.5,0.3])
axes2.set_title('Smaller axis')
```

<i>Explanation</i><br>
I have set the first axis as a normal plane starting from 0%,0% and ends at 100%,100%<br>
let us only consider the second axes for the sake of understanding <br>
it has starting point from bottom as 0.3, which means it will start from an x-axis point of 30% or in this case from 0.3 of it.<br>
similarly the y coordinate starts from 50% of axis one which is 0.5 in this case.<br>
axis 2 has width 50% , in this case 0.5, hence it will end at 0.8 or 80% of the original.

```python
#lets now plot on those axis
fig=plt.figure()

axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2=fig.add_axes([0.3,0.5,0.5,0.3])
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
#we can also plot multiple subplots using axis
fig,axes = plt.subplots(nrows=1,ncols=2)

#to fix overlapping we can use
plt.tight_layout()

#add data using a for loop to iterate between the axes
for axes_i in axes:
    axes_i.plot(x,y)    
```

In case of multiple rows and columns the above code will run into error as array of plots will be of form [[1,1],[1,2],[2,1],[2,2]].<br>
hence we need to use a dual loop to iterate further.
Example

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

#to fix overlapping we can use
plt.tight_layout()

for axes_i in axes:
    for axes_y in axes_i:
        axes_y.plot(x,y)
```

## Changing Size and DPI

```python
fig=plt.figure(figsize=(2,2))
#here we specified the max dimensions of figure using 
#figsize(width in inches,height in inches)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

To add legend follow below and add label to every plot

```python
fig=plt.figure(figsize=(4,3))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend(loc='upper right')
#defualt is "best"
```

```python
fig=plt.figure(figsize=(5,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#00FFFF',linewidth=3,alpha=0.7,linestyle='-.',
        marker='o',markersize='15',markerfacecolor='red',markeredgecolor='green')

ax.set_xlim([0,10])
ax.set_ylim([0,80])
#sets the limits of both axis
```
