```python
#answer
sns.jointplot(x='fare',y='age',data=titanic,kind='scatter')
```

```python
#answer 
sns.distplot(titanic['fare'],kde=False,color='red',bins=30)
#kde is kernel density which we dont need for this
# bins 30 on trial and error
plt.xlim([0,600])

#plt.xlim([0,600]) sets the limits for the x-axis 
```

```python
#answer
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```
